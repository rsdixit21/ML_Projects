```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/sample_data/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/sample_data/tmdb_5000_credits.csv') 
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies.shape, credits.shape
```

```python
#merging both dataFrames
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
# extracting only following columns
# movie_id
# title
# overview
# genres
# keywords
# cast
# crew
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# making string literal(list) into list
import ast
def convert(string_list):
    S = []
    for i in ast.literal_eval(string_list):
        S.append(i['name']) 
    return S 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies
```

```python
def convert2(text):
    S = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            S.append(i['name'])
        counter+=1
    return S
```

```python
def get_director(text):
    D = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            D.append(i['name'])
    return D
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def removingSpaces(L):
    S = []
    for i in L:
        S.append(i.replace(" ",""))
    return S
```

```python
movies['cast'] = movies['cast'].apply(removingSpaces)
movies['crew'] = movies['crew'].apply(removingSpaces)
movies['genres'] = movies['genres'].apply(removingSpaces)
movies['keywords'] = movies['keywords'].apply(removingSpaces)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
# Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
vector = cv.fit_transform(new_df['tags']).toarray()    
```

```python
vector.shape
```

```python
# cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie_title):
    index = new_df[new_df['title'] == movie_title].index[0]
    dist = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in dist[1:6]:
        print(new_df.iloc[i[0]].title)
        
    
```

```python
recommend('Ip Man 3')
```
